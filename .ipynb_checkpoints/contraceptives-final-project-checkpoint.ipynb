## Data Description

This dataset consists of two tables:

1. `participant` provides information on a participant’s demographics and contraceptive method choice.
2. `family_background` contains information on a participant’s husband’s education and other demographics.

A description of each table's variables is provided below:

`participant`:
* `Participant ID`: unique ID for each woman who participated
* `Wife age`: age of participant (years)
* `Wife education`: level of education (1=low, 2, 3, 4=high)
* `Wife religion`: 0=Non-Islam, 1=Islam
* `Wife working`: 0=Yes, 1=No
* `Num children`: number of children ever born
* `Contraceptive method`: participant’s choice of contraceptive method (0=no use of contraceptives, 1=use of contraceptives)

`family_background`:
* `Wife ID`: participant ID for each woman in the `participant` table
* `Husband education`: level of husband’s education (1=low, 2, 3, 4=high)
* `Husband occupation`: husband’s occupation based on predefined categories (1=professional, technical, or clerical work, 2=sales or services, 3=manual, 4=agriculture)
* `Standard-of-living`: standard-of-living index (1=low, 2, 3, 4=high)
* `Media exposure`: measures exposure and access to television, newspapers, radio, etc. (0=good, 1=not good)

## Introduction

To measure the effectiveness of Indonesia’s National Family Planning Program and subsequently improve the program, the National Indonesia Contraceptive Prevalence Survey (NICPS) collected data on fertility and family planning in 1987. 
The survey was designed to simultaneously measure contraceptive prevalence and fertility, and the factors that influence these outcomes, such as education and availability of contraceptives. We used the data from this survey, retrieved from the UCI Machine Learning Repository, for our final project. 

The participants in the survey were married women between the ages of 15-49, that were either not pregnant or didn’t know they were at the time of the interview. Although 11,884 women were interviewed, the dataset we explored is a random subset of 999 of these participants. Each row in our dataset represents one of these women through a unique ID. 

Our dataset is divided into two tables: `participant` and `family_background`. The `participant` table provides information on a participant’s demographics, like education level, religion, etc.  and contraceptive method choice. The key variable here is `Wife age`, which gives us the ages of all the participants in the survey. Data on the number of children (`Num children`) ever born to each participant is also of interest to us. Both of these variables are quantitative.

The family background table provides information on a woman’s husband’s characteristics along with other demographics like media exposure and standard of living. The variables we explore further are `Husband education` and `Husband occupation`. `Husband education` gives us the level of each participant’s husband’s education from 1 to 4, 1 being low and 4 being high. `Husband occupation` describes the occupation the husband is involved in based on predefined categories, 1 being professional, technical or clerical work; 2 being sales/services; 3 being manual labor and 4 being agriculture. Both of these variables are categorical. 

## Hypothesis Testing and Prediction Questions


By combining both of the tables in this dataset, we can study the relationships between a participants' demographics, husband's lifestyle, and her contraception usage. We are particularly interested in exploring how a wife's age may be associated with her use of a contraceptive. 

**In particular, we are interested in determining whether there is an association between a wife's age and her use of contraceptives.** Our null hypothesis is that there is no relationship between these two variables. Our alternative hypothesis is that there is a association between these variables, and, in particular, younger women are more likely to be using contraceptives compared to older women. 

**Relatedly, we wish to determine whether family characteristics like woman's age, number of children, and other attributes like husband education and occupation can be used to predict whether or not a woman uses a contraceptive.** We employ the K- Nearest Neighbors classifier method to accomplish this task.

## Hypothesis Testing

Next, we perform a hypothesis testing procedure **to determine whether there is an association between a participant wife's age and whether or not she uses a contraceptive.** The null hypothesis is that there is no such association, and our alternative hypothesis is that there is an association between the two variables, and in particular, younger women are more likely to be using contraceptives than older women. 

**This testing procedure corresponds to an A/B test.** We use an A/B test because we want to divide our dataset into two distributions, and study the difference in features between those distributions. We divide our dataset into two groups: women who do use contraceptives and women who don't. **Our test statistic will be the difference in mean ages of the wife in the two groups.** We choose this statistic because larger values of the test statistic will favor the alternative hypothesis. We employ a significance cutoff of 1%.

```python
#visualizing our distribution of wife ages grouped by contraceptive metho

contraceptive_and_age = full_data.select("Contraceptive method", "Wife age")
contraceptive_and_age.hist("Wife age", group = "Contraceptive method")
```

```python
# set the random seed so that results are reproducible
np.random.seed(1231)

#observed difference of mean wife ages between the two groups

average_tbl = contraceptive_and_age.group("Contraceptive method", np.average)
averages = average_tbl.column(1)
observed_difference = averages.item(0) - averages.item(1)
observed_difference

print ("Our observed difference is" + " " + str(observed_difference))
```

```python
#one value of our statistic under null using a permutation test

np.random.seed(1231)

ages = contraceptive_and_age.select("Wife age")
shuffled_ages = ages.sample(with_replacement = False).column(0)
original_and_shuffled = contraceptive_and_age.with_column("Shuffled wife age", shuffled_ages)
shuffled_tbl = original_and_shuffled.group("Contraceptive method", np.average)
value_1_statistic = shuffled_tbl.column(2).item(0) - shuffled_tbl.column(2).item(1)

shuffled_tbl
value_1_statistic
print("One value of our statistic under the null is" + " " + str(value_1_statistic))
```

```python
#we use iteration to simulate 5000 values for our statistic and visualise it using a histogram

np.random.seed(1231)

class_labels = full_data.select("Contraceptive method")
differences = make_array()

repititions = 5000
for i in np.arange(repititions):
    shuffled_ages = ages.sample(with_replacement = False).column(0)
    shuffled_table = class_labels.with_column("Shuffled age", shuffled_ages)
    shuffle_averages = shuffled_table.group("Contraceptive method", np.average).column(1)
    new_difference = shuffle_averages.item(0) - shuffle_averages.item(1)
    differences = np.append(differences, new_difference)
    
Table().with_column("Difference between averages", differences).hist()
plots.scatter(observed_difference, 0, color='red', s=50);
plots.title("Distribution difference of mean ages under the null");
```

Plotting our observed difference on this histogram, we see that the point lies away from the centre in the right tail of the plot. **We can reasonably conclude that our observed difference is not due to random chance, and our alternative hypothesis is true.** To quantify this conclusion, we calculate the **p-value of our experiment** in the following cell. We use a p-value cutoff of 1%. 

```python
np.random.seed(1231)

#p value
p_value = np.count_nonzero(differences >= observed_difference) / 5000
p_value

print("The p-value for our experiment is" + " " + str(p_value))
```

**Interpretation of results:**

Thus, our calculated p-value for the experiment is 0.004 or 0.4%. Using a p-value cutoff of 1%, we can reasonably reject the null hypothesis, and conclude that our experiment favors the alternative hypothesis. Thus, from our hypothesis testing, we can conclude that their is a association between the age of a woman and whether or not she uses a contraceptive. Tying these results to the plot we built in our exploration section, we can claim that in particular, younger women are more likely to be using contraceptives compared to older women. 

However, we need to be careful while interpreting the results of our experiment, since we saw earlier that our random subset consists of more participants in younger age groups compared to older age groups, which might be reason why our experiment favors the alternative hypothesis, i.e., younger women are more likely to be using contraceptives compared to older women.

## Prediction

In the following cells, we build a K- nearest neighbors (k-NN) classifier to see if we can predict whether or not a woman uses contraceptives using her age (`Wife age`) and the number of children (` Num children`). We also use characteristics of the husbands of the participants (` Husband education` and ` Husband occupation`) to be able to make more accurate predictions. 

In this case, we work with a k-NN classifier as our prediction method over linear regression because we are trying to predict an outcome for a new individual X, by using the outcomes from individuals in the dataset who are like that individual X. Our aim with this classifier is to be able to take new individual sampled from the population, pass their attributes through our classifier, and get a prediction of whether or not they use a contraceptive based on the individuals included in our dataset. Moreover, since we are handling categorical data, classification is the more appropriate choice over linear regression.

```python
#In this cell, we define the functions that evaluate the distance between a participant and the other rows
#And then, we find the k-nearest neighbors from the participant in our dataset

np.random.seed(1231)

def distance(point1, point2): 
    return np.sqrt(np.sum((point1 - point2) **2))

def all_distances(training, participant):
    attributes = training.select(" Husband education", " Husband occupation", "Wife age", " Num children")
    def row_distances(row):
        return distance(np.array(list(participant)), np.array(list(row)))
    return attributes.apply(row_distances)

def augmented_table(training, participant):
    return training.with_column('Distance', all_distances(training, participant))
    
def nearest_neighbors(training, participant, k): 
    with_dists = augmented_table(training, participant)
    sorted_by_distance = with_dists.sort('Distance')
    topk = sorted_by_distance.take(np.arange(k))
    return topk

participant = full_data.select(" Husband education", " Husband occupation", "Wife age", " Num children").row(9)
nearest_neighbors(full_data.remove(9), participant, 11).show()
```

```python
#In this cell, we define the functions that find the majority among the k-nearest neighbors
#and then classify a given participant into 0 or 1. 

np.random.seed(1231)

def majority_prediction(topneighbors):
    """ This function takes in a table consisting of the k-nearest neighbors, and returns the majority value"""
    ones = topneighbors.where('Contraceptive method', are.equal_to(1)).num_rows
    zeros = topneighbors.where('Contraceptive method', are.equal_to(0)).num_rows
    if ones > zeros:
        return 1
    else:
        return 0

def classify(training, participant, k):
    """ This function uses the 'nearest_neighbors' and 'majority' function 
        to classify a given participant into 0 or 1"""
    closestk = nearest_neighbors(training, participant, k)
    topkneighbors = closestk.select('Contraceptive method')
    return majority_prediction(topkneighbors)

classify(full_data, participant, 11)
```

```python
#in this cell, we evaluate the accuracy of our classifier
#we do this by diving the data into a training and a test set
#and then applying our classifier trained on the training set, to our test set.

np.random.seed(1231)

training_prop = 80/100

num_participants = full_data.num_rows
num_training = int(num_participants * training_prop)

shuffled_contraceptives = full_data.sample(with_replacement = False) 
training_set = shuffled_contraceptives.take(np.arange(num_training))
test_set  = shuffled_contraceptives.take(np.arange(num_training, num_participants))

def zero_count(arr):
    return len(arr) - np.count_nonzero(arr)

def equal_count(arr1, arr2):
    return zero_count(arr1 - arr2)

def accuracy(training, test, k):
    test_attributes = test.select(" Husband education", " Husband occupation", "Wife age", " Num children")
    def test_classification(row):
        return classify(training, row, k)
    classify_test = test_attributes.apply(test_classification)
    return equal_count(classify_test, test.column('Contraceptive method')) / test.num_rows

prop_accuracy = accuracy(training_set, test_set, 11)

print("My classifier has an accuracy of" + " " + str(prop_accuracy *100) + "%")
```

**Interpretation of results:**

We see that our classifier has a somewhat low accuracy (71.5%). Thus, **we can conclude that we cannot use characteristics of the family like wife age, number of children, and husband education and occupation, to predict whether or not a woman uses a contraceptive method.** A more complex model with different attributes or k values is needed to be able to accurately predict contraceptive use. 

We must recognize that **using categorical variables lead to lower accuracy in classifiers** because there are many participants in the same categories as a particular participant but with different contraceptive use. Thus, if we had more quantitative variables in our dataset, for example, number of years of marriage, we could have built a much more accurate classifier. 

## Conclusion

Using A/B Testing for our hypothesis testing procedures, we were able to **establish an association between ages of the wives and their contraceptive use.** Using the plots we developed in our exploration, we were able to further refine the alternative hypothesis and claim that **younger women are more likely to be using contraceptives compared to older women.**

However, when we used the same variable of Wife age along with Number of children and a couple other attributes to see if we can predict whether or not a woman uses contraceptives using a k-NN classifier, we got a **fairly low accuracy (66%)** Thus, we can reasonably claim that **we should NOT use family characteristics like wife age, number of children, etc. to predict contraceptive use.**

However, our analysis has a scope for improvement, since there are a number of biases visible in our dataset. Our dataset is **just a random subset of 999 women from the actual 11,884 women that were surveyed**, which makes it difficult to generalise our results to the whole population. Also, as we noticed in one of our aggregated tables, **women of younger age groups (20s and early 30s) are overrepresented in our subset**, which is another source of bias. 

Thus, to further our analysis more accurately, **we would want to work with a larger subset** or possibly all of the survey data. Also, we would want to **include more variables in the dataset**, both quantitative and qualitative, to be able to build a more precise classifier. An example would be availability of contraceptives.
